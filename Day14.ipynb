```python
def drop_sand():
    xc = 500 - x_max
    yc = 0
    
    while True:
        yc += 1
        
        if yc == y_max-1:
            return False
        elif wall_map[yc+1][xc] != '.':
            if wall_map[yc+1][xc-1] == '.':
                xc -= 1
            elif wall_map[yc+1][xc+1] == '.':
                xc += 1
            else: 
                wall_map[yc][xc] = 'O'
                return True
        elif wall_map[yc+1][xc] == '#':
            wall_map[yc][xc] = 'O'
            return True
        
    
```
