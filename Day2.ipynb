```python
# part 2
# a = rock, b = paper, c = scissors
# X = lose, Y = draw, and Z = win. 
# 1 = rock, 2 = paper, 3 = scissors
# 0 point loss, 3 point draw, 6 point win

inputRound = [['A', 'Y'], ['B', 'X'], ['C', 'Z']]

scores = [0,3,6] 
pickScore = [[3,1,2],[1, 2, 3],[2, 3, 1]]
oppDict = {'A':0, 'B':1, 'C':2}
myDict = {'X':0, 'Y':1, 'Z':2}

def scoreRound2(rnd):
    score = 0
    for hand in rnd:
        thisRnd = scores[myDict[hand[1]]] + pickScore[oppDict[hand[0]]][myDict[hand[1]]]
        score += thisRnd
        
    return score

print(scoreRound2(inputRound))
```
