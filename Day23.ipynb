```python
class grove:
    
    def __init__(self, offsets=1000):
        self.xoff = offsets
        self.yoff = offsets
        self.xmin = offsets
        self.ymin = offsets
        self.xmax = offsets
        self.ymax = offsets
        self.num_elves = 0
        self.elves = []
        self.proposals = {}
        
        self.dirs = deque(['n', 's', 'w', 'e'])
        
        
    def add_elf(self, yc, xc):
        yactual = yc+self.yoff
        xactual = xc+self.xoff
        self.elves.append((yactual, xactual))
        self.xmax = max(self.xmax, xactual)
        self.ymax = max(self.ymax, yactual)
        self.ymin = min(self.ymin, yactual)
        self.xmin = min(self.xmin, xactual)
        self.num_elves += 1
        
    def print_grove(self):
        g = []
        for y in range(self.ymin-1, self.ymax+2):
            r = ''
            for x in range(self.xmin-1, self.xmax+2):
                if (y,x) in self.elves:
                    r += '#'
                else:
                    r += '.'
            g.append(r)
        
        for r in g:
            print(r)

    def propose_moves(self):
        for (elfy, elfx) in self.elves:
            if self.count_neighbors(elfy, elfx):
                proposed_coords = self.get_proposed_space(elfy, elfx)
                if proposed_coords != (elfy, elfx):
                    if proposed_coords not in self.proposals.keys():
                        self.proposals[proposed_coords] = [(elfy, elfx)]
                    else:
                        self.proposals[proposed_coords].append((elfy, elfx))
                
        
            
    def count_neighbors(self, yc, xc):
        neighbors = 0
        offsets = [(-1, -1), (-1,  0), (-1,  1),
                   ( 1, -1), ( 1,  0), ( 1,  1),
                   (-1, -1), ( 0, -1), ( 1, -1),
                   (-1,  1), ( 0,  1), ( 1,  1)]
        
        for (y_off, x_off) in offsets:
            if (yc+y_off, xc+x_off) in self.elves:
                neighbors += 1
                
        return neighbors
        
    def get_proposed_space(self, yc, xc):
        offsets = {'n':[(-1, -1), (-1,  0), (-1,  1)],
                   's':[( 1, -1), ( 1,  0), ( 1,  1)],
                   'w':[(-1, -1), ( 0, -1), ( 1, -1)],
                   'e':[(-1,  1), ( 0,  1), ( 1,  1)]}
        
        for d in self.dirs:
            found = False
            for (y_off, x_off) in offsets[d]:
                ypot = yc + y_off
                xpot = xc + x_off
                if (ypot, xpot) in self.elves:
                    found = True
            if not found:
                return (yc + offsets[d][1][0], xc + offsets[d][1][1])
        
        return (yc, xc)
    
    def execute_turn(self):
        
        self.propose_moves()
        
        if len(self.proposals) == 0:
            return False
        
        for proposed, moving in self.proposals.items():
            if len(moving) == 1:
                self.elves.remove(moving[0])
                self.elves.append(proposed)
                yc, xc = proposed
                self.xmax = max(self.xmax, xc)
                self.ymax = max(self.ymax, yc)
                self.ymin = min(self.ymin, yc)
                self.xmin = min(self.xmin, xc)
                
            
        self.proposals = {}
        
        self.dirs.rotate(-1)
        
        return True
    
    def calculate_empty_rect(self):
        area = (self.xmax-self.xmin+1)*(self.ymax-self.ymin+1)
        area -= self.num_elves
        return area
```
